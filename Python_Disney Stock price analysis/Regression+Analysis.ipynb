1. Interactive Media Regression

```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("Interactive_media.csv")
df = df.dropna()
df = df.set_index('Date')
df.head()
```

```python
# summarize the data
df.describe()
```

```python
df.shape
```

```python
# check the column names
df.columns
```

```python
# Create  new data frame CDF AND VIZ from existing dataframe
# New dataframe contains selected columns from old dataframe
cdf=df[['Disney', 'AMZN', 'KO', 'EA', 'FL', 'HSY', 'K', 'NKE', 'NOK', 'PM',
       'PHG', 'SBUX', 'TGT', 'TSN', 'NVDA', 'INTC', 'AMD']]
cdf.head()
```

```python

viz= cdf[['Disney', 'AMZN', 'KO', 'EA', 'FL', 'HSY', 'K', 'NKE', 'NOK', 'PM',
       'PHG', 'SBUX', 'TGT', 'TSN', 'NVDA', 'INTC', 'AMD']]
viz.hist
viz.fillna(0)
pass
```

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(cdf, alpha=0.2, figsize=(15, 15), diagonal = 'orange')
```

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
plt.scatter(train.Disney,train.AMZN, color = 'blue')
plt.xlabel("Disney")
plt.ylabel("AMZN")
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['AMZN']])
train_y = np.asanyarray(train[['Disney']])
regr.fit(train_x,train_y)

# The coefficients

print ('Coefficients : ',regr.coef_)
print ('Intercept : ',regr.intercept_)
print ('R-Square :', regr.score(train_x, train_y))
```

```python
train_y_ = regr.predict(train_x)
plt.scatter(train.Disney,train.AMZN, color ='blue')
plt.plot(train_x, train_y_, color='black', linewidth =3)
```

```python

test_x = np.asanyarray(test[['Disney']])
test_y = np.asanyarray(test[['AMZN']])
test_y_ = regr.predict(test_x)

print ("Residual Sum of Squares : %.2f"
      % np.mean((test_y_ - test_y)**2))

# Explained variance score : 1 is perfect prediction

print('Variance Score : %.2f' % regr.score(test_x,test_y))
```

```python

plt.scatter(test_x,test_y, color ='blue')
plt.plot(test_x, test_y_, color ='black', linewidth =3)
plt.xlabel("Disney")
plt.ylabel("AMZN")
plt.show()
```

```python

#Non-Linear Model 
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

model = make_pipeline(PolynomialFeatures(2),Ridge())
model.fit(train_x,train_y)
train_y_ = model.predict(train_x)
plt.scatter(test_x,test_y, color = 'blue')
plt.scatter(train_x,train_y_,linewidth = 2)
```

```python


#Multiple Linear Regression Model
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['AMZN', 'KO', 'EA', 'FL', 'HSY', 'K', 'NKE', 'NOK', 'PM',
       'PHG', 'SBUX', 'TGT', 'TSN', 'NVDA', 'INTC', 'AMD']])
y = np.asanyarray(train[['Disney']])
regr.fit (x, y)
predictions = model.predict(y)

#  The Coefficients

print ('Coefficients : ',regr.coef_)
print ('R-Square :', regr.score(x, y))

```

```python
#Multiple Linear Regression Model
y_=regr.predict(test[['AMZN', 'KO', 'EA', 'FL', 'HSY', 'K', 'NKE', 'NOK', 'PM',
       'PHG', 'SBUX', 'TGT', 'TSN', 'NVDA', 'INTC', 'AMD']])
x = np.asanyarray(test[['AMZN', 'KO', 'EA', 'FL', 'HSY', 'K', 'NKE', 'NOK', 'PM',
       'PHG', 'SBUX', 'TGT', 'TSN', 'NVDA', 'INTC', 'AMD']])
y = np.asanyarray(test[['Disney']])
print("Residual Sum of Squares : %.2f"
     %np.mean((y_ -y)**2))

#Explained Variance Score : 1 is perfect prediction

print('Variance Score : %.2f' % regr.score(x,y))
```

```python
import statsmodels.api as sm
model = sm.OLS(y,x).fit()
predictions = model.predict(x)
model.summary()
```

```python
#Multiple Linear Regression Model
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['AMZN', 'KO', 'EA', 'FL', 'HSY', 'K', 'NKE', 'PM',
       'PHG', 'SBUX', 'AMD']])
y = np.asanyarray(train[['Disney']])
regr.fit (x, y)


#  The Coefficients

print ('Intercept : ',regr.intercept_)
print ('Coefficients : ',regr.coef_)
print ('R-Square :', regr.score(x, y))
```
